```python
data['Platform'].value_counts().plot.bar(color='r')
```

```python
sales=null[['Year','NA_Sales','EU_Sales','JP_Sales','Other_Sales']].groupby('Year').sum()
sales.index=sales.index.astype(int)
sales.plot.bar(figsize=(20,10),fontsize=20)
plt.tight_layout()
plt.show()
```

```python
data['Genre'].value_counts().plot.bar(color='b')
```

```python
data.groupby('Genre')[['NA_Sales','Global_Sales']].sum().plot.bar()
plt.title('DISTRIBUTION OF GENRE BY JA & NA SALES')
plt.ylabel('SALES')
plt.show()
```

```python
maps=null[['Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
maps =maps.groupby('Genre').sum()
plt.figure(figsize=(10,6))
sns.heatmap(maps,annot=True,fmt= '.0f')
```

```python
maps.plot.bar(figsize=(20,10),fontsize=30)
plt.tight_layout()
plt.show()
```

```python
null.groupby('Genre')[['Global_Sales']].sum().plot.bar(color='brown')
plt.title('DISTRIBUTION OF GLOBAL_SALES BY GENRE')
plt.ylabel('Sales')
plt.figure(figsize=(20,10))
plt.tight_layout()
```
